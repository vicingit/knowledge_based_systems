```python
# List of columns to drop
columns_to_drop = ["Code", "12m Low", "12m High", "Previous", "Adjusted Price"]

# Drop the specified columns from the DataFrame
data = df.drop(columns=columns_to_drop)
```

 FILTERING TO REMAIN WITH 2O COMPANIES

```python
# List of 20 company names you want to keep
companies_to_keep = [
    "Eaagads Ltd",
    "Kakuzi Plc",
    "Kapchorua Tea Kenya Plc",
    "Limuru Tea Plc",
    "Car and General (K) Ltd",
    "ABSA Bank Kenya Plc",
    "Co-operative Bank of Kenya Ltd",
    "Eveready East Africa Ltd",
    "Sameer Africa Plc",
    "Standard Group Plc",
    "Uchumi Supermarket Plc",
    "Total Kenya Ltd",
    "Sanlam Kenya Plc",
    "Equity Group Holdings Plc",
    "Standard Chartered Bank Kenya Ltd",
    "Kenya Airways Plc",
    "Longhorn Publishers Plc",
    "Nation Media Group Plc",
    "Crown Paints Kenya Plc",
    "Safaricom Plc"

]

# Filter the DataFrame to retain only the rows for the specified companies
filtered_data = data[data['Name'].isin(companies_to_keep)]
```

```python

# Assuming your filtered data is stored in the variable filtered_data
change_data = filtered_data['Change']

#width of plot

plt.figure(figsize=(10,6))

# Create a histogram for the "Change" column
plt.hist(change_data, bins=20, edgecolor='k')
plt.title('Price Change Distribution')
plt.xlabel('Change')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python


# Create a bar plot to compare companies based on "Day Price"
plt.figure(figsize=(12, 6))  # Set the figure size

# Sort the data by "Day Price" in descending order for better visualization
filtered_data = filtered_data.sort_values(by='Day Price', ascending=False)

# Define the x and y values for the bar plot
x_values = filtered_data['Name']
y_values = filtered_data['Day Price']

# Create the bar plot
plt.bar(x_values, y_values, color='royalblue')
plt.xlabel('Company Name')
plt.ylabel('Day Price')
plt.title('Comparison of Companies Based on Day Price')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability

plt.yticks(range(0, int(filtered_data['Day Price'].max()) + 1, 10))  # Adjust the range and step as needed
# Show the plot
plt.tight_layout()
plt.show()
```
